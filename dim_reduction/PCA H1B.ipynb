```python
# encoding 
cleanup_nums = { "CASE_STATUS":{'CERTIFIED': 1, 'DENIED': 2,'WITHDRAWN':3,'CERTIFIED-WITHDRAWN':4},
                "WILLFUL_VIOLATOR":{'N':0, 'nan':-1, 'Y':1},
                "FULL_TIME_POSITION":{'N':0, 'nan':-1, 'Y':1},
                "AGENT_REPRESENTING_EMPLOYER": {'N':0, 'nan':-1, 'Y':1}
               }
subset_df.replace(cleanup_nums, inplace=True)
```

```python
# load class labels from column 1
case_status = subset_df['CASE_STATUS']
features=subset_df.loc[:, 'AGENT_REPRESENTING_EMPLOYER':'WILLFUL_VIOLATOR']
```

```python
# printing some general information about the data
print('\ntotal number of samples (rows):', case_status.shape[0])
print('total number of features (columns):', features.shape[1])
```
