```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
from sklearn import tree
import graphviz
```

```python
appl_prods = pd.read_csv("/Users/krishna/Desktop/venkat_krishna_turlapati/training_material/appliedaicourse/datasets/apple_prod_with_screen_201901.csv")
X = appl_prods[['screen_size_in','storage_gb','cost_dollar']]
y = appl_prods[['product']]

print(X.head())
print(y.head())
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
clf.n_features_
```

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("appl_prod") 
```

```python
feature_names=list(X.columns)
print(feature_names)
target_names=list(y.values.flatten())
```

```python
dot_data = tree.export_graphviz(clf, 
                    out_file=None, 
                    feature_names=['screen_size_in', 'storage_gb', 'cost_dollar'],  
                    class_names=target_names,  
                    filled=True, 
                    rounded=True,  
                    special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
