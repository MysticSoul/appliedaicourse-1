```python
from scipy.linalg import svd
from numpy import array
from numpy import diag
from numpy import dot
from numpy import zeros
from scipy.linalg import svd
import numpy as np
```

```python
# define a matrix
A = array([[1, 2], [3, 4], [5, 6]])
print("***** Matrix A ***** ")
print(A)
print("***** Matrix SVD ***** ")
# SVD
U, s, VT = svd(A)
print("***** SVD - U ***** ")
print(U)
print("***** SVD - Sigma ***** ")
print(s)
print("***** SVD - V^T ***** ")
print(VT)
```

```python
# Reconstruct SVD


# define a matrix
A = array([[1, 2], [3, 4], [5, 6]])
print(A)
# Singular-value decomposition
U, s, VT = svd(A)
# create m x n Sigma matrix
Sigma = zeros((A.shape[0], A.shape[1]))
# populate Sigma with n x n diagonal matrix
Sigma[:A.shape[1], :A.shape[1]] = diag(s)
# reconstruct matrix
B = U.dot(Sigma.dot(VT))
print(B)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

from PIL import Image

img = Image.open('/Users/krishna/Desktop/venkat_krishna_turlapati/training_material/appliedaicourse/datasets/ellsworth.jpg')
imggray = img.convert('LA')
plt.figure(figsize=(9, 6))
plt.imshow(imggray);
```

```python
# Now I'll convert the image data into a numpy matrix, plotting the result to show the data is unchanged.
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray')
```

```python
U, sigma, V = np.linalg.svd(imgmat)
```

```python

for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
