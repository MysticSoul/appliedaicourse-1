**Distance between two points in 2d/3d/Nd**
* Distance between two points in 2 dimensions 

Let $p_1 = (x_1,y_1)$ and $p_2 = (x_2,y_2)$ be two points in cartesian coordinate system. Let d be the straight line/distance between $p_1$ and $p_2$. Then the distance can be calculated as follows 

$$d(p_2,p_1)=\sqrt {(x_2-x_1)^2 + (y_2-y_1)^2)}$$

![scatter_plot_iris](../images/2d_dis.gif)

The above distance is also called as Eucledian norm or Eucledian distance or $L_2$ distance. 

* Distance between two points in 3 dimensions

Let $p_1 = (x_1,y_1,z_1)$, $p_2 = (x_2,y_2,z_2)$ be two points in $R^3$. Let d be the straight line distance between $p_1$ and $p_2$. Then the distance can be calculated as follows 

$$d(p_2,p_1)=\sqrt {(x_2-x_1)^2 + (y_2-y_1)^2) + (z_2-z_1^2)}$$

* Distance between two points in N dimensions

Let $p = (p_1,p_2,p_3,...,p_n)$, $q = (q_1,q_2,q_3,....,q_n)$ be two points in $R^n$ . Let d be the straight line/distance between $p$ and $q$. Then the distance can be calculated as follows 

$$d(p,q)=\sqrt {(p_2-q_1)^2 + (p_2-q_2)^2 + (p_2-q_3)^2 + .... + (p_n-q_n)^2} \ \ =  \sum_{i=1}^n \sqrt {(p_i-q_i)^2} $$

**Equation of Straight Line/Plane/HyperPlane**

* Equation of a straight line in $R^2$: $ y = mx + c$
![scatter_plot_iris](../images/eq_2d_line.png)


The value $m$ is called slope or gradient of the line. The value $c$ is the y intercept. If $c =0$ then the line passes through origin.  


Alternate notation, this is also an equation for a straight line $w_1x_1 + w_2x_2 +w_0 =0$, where $y=x_2, m = -\frac {w_1} {w_2}, c = - \frac {w_0} {w_2}$  or, $\sum_{i=1}^2 w_ix_i + w_0 =0$ 


* Equation of a plane $R^3$

In three dimensions we get a plane, $w_1x_1 + w_2x_2 + w_3x_3+w_0 =0$ or, $\sum_{i=1}^3 w_ix_i + w_0 =0$

* Equation of a HyperPlane $R^3$

In $N$ dimensions, it is still a plane but its called hyper plane, $w_1x_1 + w_2x_2 +.....w_nx_n +w_0 =0$ or, $\sum_{i=1}^n w_ix_i + w_0 =0$

* Which dimension doest the intercept $c$ belongs to in higher dimensions? 

The constant $c$ is the sum of intercepts across all dimensions

**Equation of Circle/Sphere/HyperSphere**

Eqaution of a circle: $ (x-h)^2 + (y-k)^2 = r^2$, where r is the radius of the circle. 

![scatter_plot_iris](../images/eq_circle.png)

A circle in 3 dimensions becomes a sphere 

Equation of a sphere $(x - a)^2 + (y - b)^2 + (z - c)^2 = r^2$, where $(a, b, c)$ represents the center of the sphere, $r$ represents the radius, and $x$, $y$, and $z$ are the coordinates of the points on the surface of the sphere.

A sphere in n-dimensions becomes a HyperSphere

The general equation of a hypersphere is ${x_1}^2 + {x_2}^2 +{x_3}^2 ... +{x_n}^2 = r^2 $.
